DESAFIO 35: Desenvolva um programa que leia o comprimento de três retas e diga ao usuário se elas podem ou não formar um triângulo.

```
# VERIFICANDO SE AS RETAS FORMAM UM TRIÂNGULO

# =-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
# CONDIÇÃO DE EXXISTÊNCIA DE UM TRIÂNGULO
# =-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
# | b - c | < a < b + c
# | a - c | < b < a + c
# | a - b | < c < a + b 

a = float(input('digite o comprimento da rela 1: '))
b = float(input('digite o comprimento da rela 2: '))
c = float(input('digite o comprimento da rela 3: '))
if ( b - c) < a < (b + c) and (a - c) < b < (a + c) and (a - b) < c < (a + b ):
    print ('as medidas fornecidas formam um triângulo!')
else:
    print('NÃO é possível formar um triângulo!')
```
